```python
import pandas as pd
```

```python
data = pd.read_csv('/content/bank-full.csv')  ##read data
```

```python
TheseColumns = ['job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week', 'poutcome', 'subscribed']
```

```python
data[TheseColumns]  ##check unique values
```

```python
for i in TheseColumns:  ##encode categorical data (convert from categorical to numerical)
  Dict = {i:j for i,j in zip(data[i].unique(),range(data[i].nunique()))}
  data[f'Enc_{i}'] = data[i].map(Dict)
  data.drop([i],axis=1,inplace=True)
```

```python
data  ##check final data
```

```python
from sklearn.cluster import KMeans  ##import kmeans
```

```python
KMeansModel = KMeans(n_clusters=5, init = 'k-means++', random_state = 33, algorithm='auto')  ##create model
```

```python
KMeansModel.fit(data)  ##fit model
```

```python
pd.DataFrame(KMeansModel.cluster_centers_,columns=data.columns, index=['ClusterA','ClusterB','ClusterC','ClusterD','ClusterE'])  ##check cluster centers
```

```python
len(KMeansModel.labels_)  ##check labels
```

```python
pd.Series(KMeansModel.labels_).value_counts()  ##check labels counts
```

```python
KMeansModel.inertia_  ##check inertia
```

```python
from tqdm import tqdm  ##check intertia for different clusters
intertiaDict = {}
for i in tqdm(range(2,16)):
  KMeansModel = KMeans(n_clusters=i, init = 'k-means++', random_state = 33, algorithm='auto')
  KMeansModel.fit(data)
  intertiaDict[i] = KMeansModel.inertia_
```

```python
import matplotlib.pyplot as plt  ##plot intertia for different clusters
plt.figure(figsize=(7,4))
plt.title('ElbowMethod')
plt.xlabel('Number of Clusters')
plt.ylabel('SSD')
XGraph = intertiaDict.keys()
YGraph = intertiaDict.values()
plt.plot(XGraph,YGraph, linewidth = 1, color = 'r', alpha = 1, linestyle = 'solid')
```
